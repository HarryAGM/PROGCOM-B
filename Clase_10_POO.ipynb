```python
class Casa:
  def __init__(self,color,num_habitaciones,num_baos,tiene_patio,tiene_parqueadero):
    """
    Constructor de la clase casa
    Argumentos:
      color(str): definir el color de la casa
      num_habitaciones(int): nmero de habitaciones
      num_baos(int):nmero de baos
      tiene_patio(bool): si la casa posee patio
      tiene_parqueadero(bool): si la casa tiene parqueadero
    """
    self.color=color
    self.num_habitaciones=num_habitaciones
    self.num_baos=num_baos
    self.tiene_patio=tiene_patio
    self.tiene_parqueadero=tiene_parqueadero

  def describir(self):
    """
    Mtodo para describir la casa
    """
    descripcion=f"La casa es de color {self.color}. Tiene {self.num_habitaciones} habitacion(es) y tiene {self.num_baos} bao(s)."

    if self.tiene_patio:
      descripcion+="\nLa casa posee un patio."
    if self.tiene_parqueadero:
      descripcion+="\nLa casa posee un parqueadero."
    return descripcion

  def pintar(self,nuevo_color):
    """
    Mtodo para pintar la casa
    Argumentos:
      nuevo_color(str): nuevo color de la casa
    """
    self.color=nuevo_color
    print(f"La casa ha sido pintada de color {self.color}.")
```

```python
mi_casa=Casa("roja con blanco", 5,2,True,True)
type(mi_casa)
```

```python
print(mi_casa.describir())
```

```python
class Floristeria(Casa):#Herencia
  def __init__(self,color,num_habitaciones,num_baos,tiene_patio,tiene_parqueadero,nombre,flores):
    """
    Constructor de la clase floristeria
    Argumentos:
      color(str): color de la floristeria
      num_habitaciones(int): nmero de habitaciones
      num_baos(int): nmero de baos
      tiene_patio(bool): si la floristeria posee un patio
      tiene_parqueadero(bool): si la floristeria posee un parqueadero
      nombre(str): nombre de la floristeria
      flores(list): lista de flores
    """
    super().__init__(color,num_habitaciones,num_baos,tiene_patio,tiene_parqueadero)
    self.nombre=nombre
    self.flores=flores

  def describir_floristeria(self):
    descripcion_casa=super().describir()
    descripcion_floristeria=f"\nLa floristeria se llama {self.nombre} y vende las siguientes flores: {', '.join(self.flores)}."
    return descripcion_casa+descripcion_floristeria
```

```python
mi_floristeria=Floristeria("amarilla",1,1,False,False,"Floristeria",["tulipn","rosas","girasoles"])
```

```python
mi_floristeria.pintar("rosa")
```
