```
class Flowers:
    def __init__(self, titulo, color_puerta, marco_puerta, manija_puerta, tam_puerta,
                 girasoles, pasto_frontal, parte_derecha, parte_izquierda,
                 parte_trasera, base, techo):
        """
        Constructor de la clase Flowers.
        Argumentos:
            titulo (dict): Información del título frontal (texto, color, altura, tamaño).
            color_puerta (str): Color principal de la puerta.
            marco_puerta (str): Color del marco de la puerta.
            manija_puerta (str): Color de la manija de la puerta.
            tam_puerta (str): Tamaño de la puerta.
            girasoles (list): Lista de girasoles con altura, color y posición.
            pasto_frontal (dict): Color y altura del pasto del frente.
            parte_derecha (dict): Información de la cara derecha.
            parte_izquierda (dict): Información de la cara izquierda.
            parte_trasera (dict): Información de la cara trasera.
            base (dict): Información de la base de la casita.
            techo (dict): Información del techo.
        """
        self.titulo = titulo
        self.color_puerta = color_puerta
        self.marco_puerta = marco_puerta
        self.manija_puerta = manija_puerta
        self.tam_puerta = tam_puerta
        self.girasoles = girasoles
        self.pasto_frontal = pasto_frontal
        self.parte_derecha = parte_derecha
        self.parte_izquierda = parte_izquierda
        self.parte_trasera = parte_trasera
        self.base = base
        self.techo = techo

    def describir_frontal(self):
        texto = (
            f"PARTE FRONTAL\n\n"
            f"Título: {self.titulo['Texto']} ({self.titulo['Color']}), a {self.titulo['Altura']} desde la base.\n"
            f"Tamaño del título: {self.titulo['Tamaño']}.\n\n"
            f"Puerta de color {self.color_puerta} con marco {self.marco_puerta} y manija {self.manija_puerta} "
            f"({self.tam_puerta}).\n\n"
            f"Girasoles: uno a la {self.girasoles[0]['Lado']} ({self.girasoles[0]['Altura']} de alto) "
            f"y otro a la {self.girasoles[1]['Lado']} ({self.girasoles[1]['Altura']} de alto), ambos "
            f"de color {self.girasoles[0]['Color']} con centro {self.girasoles[0]['Centro']}.\n\n"
            f"Pasto {self.pasto_frontal['Color']} de {self.pasto_frontal['Altura']} de altura."
        )
        return texto

    def describir_derecha(self):
        p = self.parte_derecha
        return (
            f"PARTE DERECHA\n\n"
            f"Dimensiones: {p['Dimensiones']}.\n"
            f"Flor morada con {p['Flor morada']['Pétalos']} pétalos, centro {p['Flor morada']['Centro']} "
            f"y color {p['Flor morada']['Color']} ({p['Flor morada']['Tamaño']}).\n\n"
            f"Flores rojas: {p['Flores rojas']['Cantidad']} en total, tallos {p['Flores rojas']['Tallo']}.\n\n"
            f"Pasto de {p['Pasto']['Altura']} de altura."
        )

    def describir_izquierda(self):
        p = self.parte_izquierda
        return (
            f"PARTE IZQUIERDA\n\n"
            f"Ventana de marco {p['Ventana']['Marco']} ({p['Ventana']['Tamaño del marco']}).\n"
            f"En su interior hay un {p['Ventana']['Flor interior']['Tipo']} "
            f"{p['Ventana']['Flor interior']['Color']} de {p['Ventana']['Flor interior']['Altura']}.\n\n"
            f"Afuera hay {p['Tulipanes exteriores']['Cantidad']} tulipanes "
            f"de {p['Tulipanes exteriores']['Altura']} de alto, color {p['Tulipanes exteriores']['Color']}.\n\n"
            f"Pasto de {p['Pasto']['Altura']} de altura."
        )

    def describir_trasera(self):
        p = self.parte_trasera
        return (
            f"PARTE TRASERA\n\n"
            f"Tulipán grande color {p['Tulipán grande']['Color']} con tallo {p['Tulipán grande']['Tallo']} "
            f"y altura máxima de {p['Tulipán grande']['Altura máxima']}.\n"
            f"Líneas decorativas {p['Tulipán grande']['Líneas decorativas']}.\n\n"
            f"Flores: 2 rojas y 2 moradas con centros contrastantes.\n"
            f"Girasoles: {p['Girasoles']}.\n"
            f"Pasto de {p['Pasto']['Altura']} de altura."
        )

    def describir_base(self):
        p = self.base
        return (
            f"BASE\n\n"
            f"Color: {p['Color']}.\n"
            f"Dimensiones: {p['Dimensiones']}.\n"
            f"Iniciales: {p['Iniciales']} en color {p['Color de iniciales']}."
        )

    def describir_techo(self):
        p = self.techo
        return (
            f"TECHO\n\n"
            f"Color: {p['Color']}.\n"
            f"Material: {p['Material']}.\n"
            f"Patrón: {p['Patrón']}.\n"
            f"Tejas: {p['Total de tejas']} en total ({p['Tejas por lado']} por lado).\n"
            f"Tamaño: {p['Tamaño']}."
        )

    def describir_todo(self):

        #descripción completa de toda la casita.
        return (
            f"{self.describir_frontal()}\n\n"
            f"{self.describir_derecha()}\n\n"
            f"{self.describir_izquierda()}\n\n"
            f"{self.describir_trasera()}\n\n"
            f"{self.describir_base()}\n\n"
            f"{self.describir_techo()}"
        )
```
